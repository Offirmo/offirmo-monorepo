### Schema

#### Custom

https://zod.dev/api?id=custom

```typescript
import * as z from 'zod'

const px = z.custom<`${number}px`>(
  val => typeof val === "string" && /^\d+px$/.test(val),
  //'Custom error',
);

type px = z.infer<typeof px>; // `${number}px`

px.parse("42px"); // "42px"
px.parse("42vw"); // throws;
```
