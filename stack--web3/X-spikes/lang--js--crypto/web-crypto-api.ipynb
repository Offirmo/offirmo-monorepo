## Keys



```typescript
crypto.subtle
```

```typescript
function getMembers(obj)
{
    const res = [];
    for(let m in obj) {
       res.push(m)
    }
    return res;
}
getMembers(crypto.subtle)
```

## Key gen



```typescript
let keyPair = await crypto.subtle.generateKey(
  {
    name: "RSA-OAEP",
    modulusLength: 4096,
    publicExponent: new Uint8Array([1, 0, 1]),
    hash: "SHA-256",
  },
  true,
  ["encrypt", "decrypt"],
);

keyPair
```
