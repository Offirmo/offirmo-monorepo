# Bucketing

## Log(s)

```typescript
function showcase(i) {
    console.log(
        i.toFixed(2).padStart(12, ' '),

        Math.log(i).toFixed(1).padStart(10, ' '),
        Math.log1p(i).toFixed(1).padStart(10, ' '),
        Math.log2(i).toFixed(1).padStart(10, ' '),
        Math.log10(i).toFixed(1).padStart(10, ' '),
    )
}

console.log(`      Amount        log      log1p       log2      log10`)
showcase(0)
for (let i = 0.1; i <= 100_000_000; i = i * 10) {
    showcase(i)
    //test(i + 0.1)
    showcase(i * 2.51)
}
```

## simple "at most" bucketing by powers of 10

```typescript
function test(i) {
    const log10 = Math.log10(i);

    //const temp1 = log10 - 0.00001
    //const temp2 = Math.max(0, 1 + Math.floor(temp1))
    //const atMost = i === 0 ? 0 : Math.floor(10 ** (temp2))

    const temp1 = Math.ceil(log10)
    const temp2 = Math.floor(10 ** temp1)
    const atMost = i === 0 ? 0 : Math.max(1, temp2);

    console.log(
        i.toFixed(2).padStart(12, ' '),

        Math.log10(i).toFixed(1).padStart(10, ' '),

        temp1.toFixed(1).padStart(10, ' '),
        temp2.toFixed(0).padStart(10, ' '),
        String(atMost).padStart(10, ' '),
    )
}

console.log(`      Amount      log10      temp1      temp2     atMost`)
test(0)
for (let i = 0.1; i <= 100_000_000; i = i * 10) {
    test(i)
    //test(i + 0.1)
    test(i * 2.51)
}
```
