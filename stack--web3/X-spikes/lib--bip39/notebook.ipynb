#### entropy = mnemonic length

> You can customize the strength of the generated mnemonic by passing a value between 128 and 256 as the second argument to the generateMnemonic function. This value must be a multiple of 32.
> 
> https://v1.viem.sh/docs/accounts/mnemonic.html#generating-mnemonics

```typescript
import * as bip39 from "npm:@scure/bip39"
import { wordlist } from "npm:@scure/bip39/wordlists/english.js"

;[128, 192, 256].forEach(entropy => {
	const phrase = bip39.generateMnemonic(wordlist, entropy)
	const words_count = phrase.split(' ').length
	console.log(`entropy=${entropy}: ${words_count} words, ex.`, phrase)
})
```
